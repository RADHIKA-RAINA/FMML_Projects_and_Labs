## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

````python
import numpy as np

# Gradient Descent Function
def gradient_descent(derivative_func, initial_guess, multiplier, precision, max_iter):
    """
    Performs gradient descent given a derivative function and parameters.

    Args:
    derivative_func (function): The derivative of the target function.
    initial_guess (float): Starting point for the gradient descent.
    multiplier (float): Learning rate (step size).
    precision (float): Convergence criterion (minimum step size).
    max_iter (int): Maximum number of iterations.

    Returns:
    list: x values at each iteration.
    """
    new_x = initial_guess
    x_list = [new_x]  # Track x values
    for _ in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient
        x_list.append(new_x)
        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    return x_list

# Define the derivative function for g(x)
def dg(x):
    return 4*x**3 - 8*x  # Derivative of g(x) = x^4 - 4x^2 + 5

# TASK-5: Run gradient descent 3 times with different initial guesses
n = 100
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                             multiplier=0.0005, precision=0.0001, max_iter=n)

mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                             multiplier=0.001, precision=0.0001, max_iter=n)

high_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                              multiplier=0.002, precision=0.0001, max_iter=n)

# TASK-6: Run gradient descent with a high multiplier to test divergence
insane_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9,
                                 multiplier=0.25, precision=0.0001, max_iter=n)

# Visualization of gradient descent paths for different learning rates
import matplotlib.pyplot as plt

x_range = np.linspace(-2, 2, 1000)
g = lambda x: x**4 - 4*x**2 + 5  # Define g(x)

plt.figure(figsize=(12, 6))

# Plot the function g(x)
plt.plot(x_range, g(x_range), label="g(x) = x^4 - 4x^2 + 5", color="blue")

# Plot gradient descent paths
plt.plot(low_gamma, [g(x) for x in low_gamma], 'o-', label="Low gamma (0.0005)", color="green")
plt.plot(mid_gamma, [g(x) for x in mid_gamma], 'o-', label="Mid gamma (0.001)", color="orange")
plt.plot(high_gamma, [g(x) for x in high_gamma], 'o-', label="High gamma (0.002)", color="red")
plt.plot(insane_gamma, [g(x) for x in insane_gamma], 'o-', label="Insane gamma (0.25)", color="purple")

# Customize plot
plt.title("Gradient Descent Paths with Different Learning Rates")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()
plt.show()
````

````python
# Function to perform gradient descent and track cost reduction
def gradient_descent_with_cost(derivative_func, cost_func, initial_guess, multiplier, precision, max_iter):
    """
    Performs gradient descent while tracking cost function values.

    Args:
    derivative_func (function): Derivative of the target function.
    cost_func (function): Target function (cost function).
    initial_guess (float): Starting point for the gradient descent.
    multiplier (float): Learning rate (step size).
    precision (float): Convergence criterion (minimum step size).
    max_iter (int): Maximum number of iterations.

    Returns:
    list: x values at each iteration.
    list: Corresponding cost function values at each iteration.
    """
    new_x = initial_guess
    x_list = [new_x]  # Track x values
    cost_list = [cost_func(new_x)]  # Track cost function values
    for _ in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient
        x_list.append(new_x)
        cost_list.append(cost_func(new_x))
        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    return x_list, cost_list

# Define the cost function and its derivative
def g(x):
    return x**4 - 4*x**2 + 5  # Cost function

def dg(x):
    return 4*x**3 - 8*x  # Derivative of the cost function

# Parameters for gradient descent
initial_guess = 3
learning_rate = 0.001
precision = 0.0001
max_iter = 100

# Perform gradient descent and track cost reduction
x_values, cost_values = gradient_descent_with_cost(derivative_func=dg, cost_func=g,
                                                   initial_guess=initial_guess,
                                                   multiplier=learning_rate,
                                                   precision=precision,
                                                   max_iter=max_iter)

# Plot reduction in cost for each iteration
plt.figure(figsize=(8, 5))
plt.plot(range(len(cost_values)), cost_values, 'o-', color='blue', label='Cost Reduction')
plt.title("Reduction in Cost per Iteration")
plt.xlabel("Iteration")
plt.ylabel("Cost Function Value (g(x))")
plt.grid()
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the functions
def f1(x, y):
    return 13 - x**2 - y**2 + 1

def f2(x, y):
    r = 3 - x**2 - y**2
    return 1 / (r + 1)

# Define gradients
def grad_f1(x, y):
    df_dx = -2 * x
    df_dy = -2 * y
    return np.array([df_dx, df_dy])

def grad_f2(x, y):
    r = 3 - x**2 - y**2
    dr_dx = -2 * x
    dr_dy = -2 * y
    df_dx = -dr_dx / (r + 1)**2
    df_dy = -dr_dy / (r + 1)**2
    return np.array([df_dx, df_dy])

# Visualize the functions
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

Z1 = f1(X, Y)
Z2 = f2(X, Y)

fig = plt.figure(figsize=(12, 6))

# Plot f1(x, y)
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z1, cmap='viridis', alpha=0.8)
ax1.set_title("3D Plot of $f_1(x, y)$")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("$f_1(x, y)$")

# Plot f2(x, y)
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z2, cmap='plasma', alpha=0.8)
ax2.set_title("3D Plot of $f_2(x, y)$")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_zlabel("$f_2(x, y)$")

plt.tight_layout()
plt.show()

# Gradient Descent Function
def gradient_descent_2d(grad_func, initial_guess, learning_rate, precision, max_iter):
    x, y = initial_guess
    points = [(x, y)]
    for _ in range(max_iter):
        grad = grad_func(x, y)
        x_new = x - learning_rate * grad[0]
        y_new = y - learning_rate * grad[1]
        points.append((x_new, y_new))
        if np.sqrt((x_new - x)**2 + (y_new - y)**2) < precision:
            break
        x, y = x_new, y_new
    return points

# Minimize f1(x, y) and f2(x, y)
initial_guess = (1.5, 1.5)
learning_rate = 0.1
precision = 0.0001
max_iter = 100

points_f1 = gradient_descent_2d(grad_f1, initial_guess, learning_rate, precision, max_iter)
points_f2 = gradient_descent_2d(grad_f2, initial_guess, learning_rate, precision, max_iter)

# Plot gradient descent paths on 3D plots
def plot_gradient_descent_3d(func, points, title, cmap):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)
    ax.plot_surface(X, Y, Z, cmap=cmap, alpha=0.8)

    # Plot the gradient descent path
    path_x, path_y = zip(*points)
    path_z = [func(x, y) for x, y in points]
    ax.plot(path_x, path_y, path_z, color='red', marker='o', label="Gradient Descent Path")

    ax.set_title(title)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("f(x, y)")
    ax.legend()
    plt.show()

plot_gradient_descent_3d(f1, points_f1, "Gradient Descent on $f_1(x, y)$", 'viridis')
plot_gradient_descent_3d(f2, points_f2, "Gradient Descent on $f_2(x, y)$", 'plasma')
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````
